### Creating cursor to communicate with DB

```python
cursor=dbi.cursor()
```

```python
#Structured class for the query
class Struct():
    def __init__(self,cur):
        rows=cursor.fetchall()
        cols = zip(*rows)
        self.header = [item[0] for item in cursor.description]
        self.count = cur.rowcount
        for k in range(len(self.header)):
            setattr(self,self.header[k],array(cols[k]))
        del rows,cols
```

## New photo-z PDF objects to access the DB

### Select the first 9 full photo-z pdf from PHOTOZ_PDF

```python
#Note the Union All statement to join at the bottom
#The zbins values
cursor.execute("""\
Select coadd_objects_id,tpz from MCARRAS2.PHOTOZ_PDF_SVA1_GOLD \
where rownum < 10 \
UNION ALL select -1,tpz_zvals from MCARRAS2.ZBINS""")
All=cursor.fetchall()
#Last rows has the zbins values
zbins=array(All[-1][1])
fig=plt.figure(figsize=(8,8))
for i in range(9):
    ax=fig.add_subplot(3,3,i+1)
    pdf=array(All[i][1]) #Note the full pdf is already in a list at index 1
    ax.plot(zbins,pdf,'k-')
    ax.set_yticks([])
    ax.set_xticks([0,0.3,0.6,0.9,1.2,1.5,1.8])
    ax.text(0.05,max(pdf)*0.9,'ID = '+str(All[i][0]),fontsize=10)
```

# Compute N(z) for the first 10000 objects within the redshift bin 0.1 < z < 0.4

```python

cursor.execute("""\
Select MCARRAS2.NZ(MCARRAS2.PHZ.TOTABLE(tpz)) from MCARRAS2.PHOTOZ_PDF_SVA1_GOLD  \
where MCARRAS2.PHZ.MEAN(tpz) BETWEEN 0.1 and 0.4 and rownum < 10000 \
UNION ALL select tpz_zvals from MCARRAS2.ZBINS""")
All=cursor.fetchall()
#2 rows only, first with Nz and second with zbins
Nz=All[0][0]
zbins=All[1][0]
plt.figure(figsize=(6,6))
plt.plot(zbins,Nz,'k-')
plt.xlabel('redshift')
plt.ylabel('N(z)') 
```

### How to get the metrics from the PDF tables

```python
cursor.execute("""\
Select MCARRAS2.PHZ.MEAN(tpz) mean, MCARRAS2.PHZ.MEDIAN(tpz) peak from \
MCARRAS2.PHOTOZ_PDF_SVA1_GOLD where rownum < 1000""")
All=cursor.fetchall()
cols=array(zip(*All))
plt.figure(figsize=(6,6))
plt.plot(cols[0],cols[1],'ko')
plt.xlabel('PDF mean')
plt.ylabel('PDF median')
```
